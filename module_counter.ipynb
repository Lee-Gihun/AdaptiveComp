```python
import math

def _target_setter(position_flops, base=0.25):
    hard_target = []
        
    for F_i in position_flops:
        target = min(math.sin(base+math.sin(F_i*math.pi/2)), 1)
        hard_target.append(target)
        
    return hard_target


_target_setter([0.35217])
```
